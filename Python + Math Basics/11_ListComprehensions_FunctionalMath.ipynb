# 11 – List Comprehensions & Functional Math

## 📖 Theory

A **list comprehension** is a concise, one-line way to create a list by applying an expression to each element of an iterable — like **set-builder notation** in mathematics:

$$\{x^2 \mid x \in \{1, 2, 3, 4, 5\}\} \Rightarrow \texttt{[x**2 for x in range(1,6)]}$$

Syntax: `[expression for variable in iterable if condition]`

**Functional tools:** `map()`, `filter()`, `reduce()` apply functions across iterables, making mathematical transformations elegant and readable.

```
# ── Basic list comprehension ──────────────────────────────────────────
# Generate squares of 1 to 10
# Traditional loop version:
squares_loop = []
for x in range(1, 11):
    squares_loop.append(x ** 2)

# List comprehension version (equivalent, shorter):
squares_comp = [x ** 2 for x in range(1, 11)]  # x² for x from 1 to 10

print('Loop version :', squares_loop)
print('Comprehension:', squares_comp)   # identical results
```

```
# ── Comprehension with condition (filtering) ─────────────────────────
# Get only even squares: {x² | x ∈ {1..20}, x even}
even_squares = [x**2 for x in range(1, 21) if x % 2 == 0]
print('Even squares :', even_squares)

# Prime number sieve using comprehension (trial division)
def is_prime(n):
    if n < 2: return False              # 0 and 1 are not prime
    for i in range(2, int(n**0.5)+1):  # check divisors up to √n
        if n % i == 0: return False    # found a divisor → not prime
    return True                         # no divisors → prime

primes = [n for n in range(2, 50) if is_prime(n)]  # primes below 50
print('Primes < 50  :', primes)
```

```
# ── map() – apply a function to every element ────────────────────────
# map(function, iterable) → applies function to each element
temperatures_c = [0, 20, 37, 100]   # Celsius temperatures

# Convert each to Fahrenheit: F = C × 9/5 + 32
to_f = lambda c: c * 9/5 + 32           # lambda function for conversion
temps_f = list(map(to_f, temperatures_c))  # apply to every element

print('Celsius   :', temperatures_c)
print('Fahrenheit:', temps_f)
```

```
# ── filter() – keep elements matching a condition ─────────────────────
numbers = list(range(-5, 11))   # [-5, -4, ..., 0, ..., 10]

# Keep only positive numbers: {x | x > 0}
positives = list(filter(lambda x: x > 0, numbers))  # filter keeps True values

# Keep only numbers divisible by 3
div_by_3 = list(filter(lambda x: x % 3 == 0, numbers))

print('All      :', numbers)
print('Positive :', positives)
print('Div by 3 :', div_by_3)
```

```
from functools import reduce   # reduce must be imported explicitly

# ── reduce() – aggregate elements into one value ─────────────────────
numbers = [1, 2, 3, 4, 5]

# Compute product: 1 × 2 × 3 × 4 × 5 = 120
product = reduce(lambda x, y: x * y, numbers)  # combine pairs left to right
print('Product of [1..5] :', product)   # 120

# Find maximum using reduce
data = [14, 3, 77, 22, 55, 8]
maximum = reduce(lambda a, b: a if a > b else b, data)  # keep the larger
print('Maximum of data   :', maximum)   # 77
```

### 💡 Key Takeaway
List comprehensions mirror set-builder notation from mathematics and make transformations readable. `map` transforms, `filter` selects, `reduce` aggregates — the three pillars of functional programming.
