# 06 – Loops & Mathematical Iteration

## 📖 Theory

**Loops** allow us to repeat a block of code multiple times — essential for:
- Computing **series** (1 + 2 + 3 + ... + n)
- **Factorials** (n! = 1 × 2 × 3 × ... × n)
- **Fibonacci sequences**, prime checking, etc.

### Two main loop types:
1. **`for` loop** – iterate over a range or collection (known number of repetitions)
2. **`while` loop** – repeat while a condition is True (unknown number of repetitions)

`range(start, stop, step)` generates a sequence of integers from `start` up to (not including) `stop`.

```
# ── for loop with range ───────────────────────────────────────────────
# range(1, 6) generates: 1, 2, 3, 4, 5
print('Counting from 1 to 5:')
for i in range(1, 6):      # i takes values 1, 2, 3, 4, 5 in order
    print(f'  i = {i}')
```

```
# ── Sum of first n natural numbers: S = n(n+1)/2 ─────────────────────
n = 10        # we want the sum of 1 + 2 + ... + 10
total = 0     # start with sum = 0

for i in range(1, n + 1):   # i goes 1, 2, 3, ..., 10
    total += i               # add i to running total (total = total + i)

# Verify using the mathematical formula: n(n+1)/2
formula_result = n * (n + 1) // 2   # integer division

print(f'Sum (loop)    : {total}')          # 55
print(f'Sum (formula) : {formula_result}') # 55 — should match!
```

```
# ── Factorial: n! = 1 × 2 × 3 × ... × n ─────────────────────────────
n = 6           # compute 6!
factorial = 1   # start with product = 1 (not 0, as we are multiplying)

for i in range(1, n + 1):   # i goes 1, 2, 3, 4, 5, 6
    factorial *= i           # multiply accumulator by i
    print(f'  Step {i}: {i}! so far = {factorial}')

print(f'\n6! = {factorial}')  # 720
```

```
# ── while loop: digits of Fibonacci sequence ─────────────────────────
# Fibonacci: each number = sum of the two before it → 0, 1, 1, 2, 3, 5, 8, ...
a, b = 0, 1       # initialise first two Fibonacci numbers
limit = 100       # print all Fibonacci numbers below 100

print('Fibonacci numbers below', limit, ':')
while a < limit:           # keep looping while a is less than limit
    print(a, end='  ')     # print current value (end=' ' keeps on same line)
    a, b = b, a + b        # update: next = current + previous (simultaneous swap)
```

```
# ── Nested loop: multiplication table ────────────────────────────────
print('Multiplication Table (1–5)\n')
print('   ', end='')              # header spacing
for col in range(1, 6):
    print(f'{col:4}', end='')     # column headers
print()                           # newline after header
print('─' * 25)

for row in range(1, 6):           # outer loop: rows 1 to 5
    print(f'{row:2} |', end='')   # row label
    for col in range(1, 6):       # inner loop: columns 1 to 5
        print(f'{row*col:4}', end='')  # compute and print product
    print()                       # newline at end of each row
```

### 💡 Key Takeaway
`for` loops are perfect for finite mathematical iterations (sums, products). `while` loops handle open-ended processes like Fibonacci generation. Nested loops solve 2D problems like multiplication tables.
