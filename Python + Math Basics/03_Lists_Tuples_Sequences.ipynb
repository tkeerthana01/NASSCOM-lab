# 03 – Lists, Tuples & Mathematical Sequences

## 📖 Theory

### List
A **list** is an **ordered, mutable** (changeable) collection of items enclosed in `[ ]`.
Think of it as a mathematical **sequence** where each element has an index starting from 0.

### Tuple
A **tuple** is an **ordered, immutable** (unchangeable) collection enclosed in `( )`.
Use tuples for data that must not change, e.g., coordinates `(x, y)`.

### Why useful in Maths?
- Store a dataset of numbers
- Represent vectors and coordinates
- Perform element-wise operations

```
# ── Creating a list of numbers ───────────────────────────────────────
scores = [88, 76, 92, 55, 67, 83, 95]  # list of exam scores

print('Scores   :', scores)            # full list
print('First    :', scores[0])         # index 0 = first element
print('Last     :', scores[-1])        # index -1 = last element
print('Length   :', len(scores))       # total number of elements
```

```
# ── List slicing ─────────────────────────────────────────────────────
numbers = [10, 20, 30, 40, 50, 60]    # a list of numbers

print(numbers[1:4])   # [20, 30, 40]  elements from index 1 up to (not including) 4
print(numbers[:3])    # [10, 20, 30]  first 3 elements
print(numbers[3:])    # [40, 50, 60]  from index 3 to end
print(numbers[::2])   # [10, 30, 50]  every 2nd element (step=2)
```

```
# ── Modifying lists ──────────────────────────────────────────────────
primes = [2, 3, 5, 7]   # first 4 prime numbers

primes.append(11)        # add 11 at the end
primes.append(13)        # add 13 at the end
print('After append:', primes)   # [2, 3, 5, 7, 11, 13]

primes.remove(3)         # remove the value 3 from the list
print('After remove:', primes)   # [2, 5, 7, 11, 13]

primes.sort()            # sort the list in ascending order
print('Sorted      :', primes)   # [2, 5, 7, 11, 13]
```

```
# ── Mathematical summary statistics from a list ──────────────────────
data = [4, 8, 15, 16, 23, 42]    # a dataset

total  = sum(data)                # sum of all elements
count  = len(data)                # how many elements
mean   = total / count            # arithmetic mean = sum / count
minval = min(data)                # smallest value
maxval = max(data)                # largest value

print(f'Data  : {data}')
print(f'Sum   : {total}')
print(f'Count : {count}')
print(f'Mean  : {mean:.2f}')
print(f'Min   : {minval}')
print(f'Max   : {maxval}')
```

```
# ── Tuples – immutable sequences ─────────────────────────────────────
point = (3, 4)           # 2D coordinate (x, y) – should not change
color = (255, 128, 0)    # RGB color values

print('Point  :', point)     # (3, 4)
print('X-coord:', point[0])  # 3
print('Y-coord:', point[1])  # 4

# Distance from origin using Pythagoras: d = √(x² + y²)
distance = (point[0]**2 + point[1]**2) ** 0.5  # 0.5 = square root
print(f'Distance from origin: {distance:.4f}')  # 5.0
```

### 💡 Key Takeaway
Lists are mutable sequences ideal for storing datasets. Tuples are immutable and perfect for mathematical coordinates. Python's built-in `sum()`, `min()`, `max()` make data analysis easy.
