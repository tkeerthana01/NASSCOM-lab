# 13 – File Handling & Data Storage for Mathematics

## 📖 Theory

In data science and mathematics, we constantly read data from files and write results back.

### Common file types:
| Format | Use case | Python tool |
|--------|----------|-----------|
| `.txt` | Plain text, raw data | built-in `open()` |
| `.csv` | Tabular/spreadsheet data | `csv` module |
| `.json` | Structured data (key-value) | `json` module |

### File modes:
- `'r'` – read (default)
- `'w'` – write (overwrites existing file)
- `'a'` – append (adds to end)
- Always use `with open(...) as f:` — it automatically closes the file

```
# ── Writing numbers to a text file ──────────────────────────────────
filename = 'math_data.txt'   # name of the file to create

# 'w' mode creates file (or overwrites if it exists)
with open(filename, 'w') as f:                    # open file for writing
    f.write('Mathematics Data File\n')            # write heading
    f.write('='*30 + '\n')                        # separator line
    for i in range(1, 11):                         # write squares 1 to 10
        f.write(f'{i:2d} squared = {i**2:4d}\n') # formatted line

print(f'File "{filename}" written successfully.')
```

```
# ── Reading from a text file ─────────────────────────────────────────
filename = 'math_data.txt'

# 'r' mode opens for reading
with open(filename, 'r') as f:    # open file for reading
    content = f.read()            # read entire file as a string

print('File contents:')
print(content)
```

```
import csv    # Python's built-in CSV (Comma-Separated Values) module

# ── Writing a CSV file ────────────────────────────────────────────────
csv_file = 'students.csv'

students = [
    ['Name',  'Math', 'Science', 'English'],   # header row
    ['Aarav',   88,      75,        82],
    ['Divya',   92,      89,        95],
    ['Ravi',    76,      80,        70],
    ['Sneha',   95,      91,        88],
]

with open(csv_file, 'w', newline='') as f:    # open for writing
    writer = csv.writer(f)                     # create a CSV writer object
    writer.writerows(students)                 # write all rows at once

print(f'CSV file "{csv_file}" created.')
```

```
import csv

# ── Reading and computing averages from CSV ───────────────────────────
csv_file = 'students.csv'

with open(csv_file, 'r') as f:
    reader = csv.DictReader(f)     # DictReader reads each row as a dictionary
    print(f'{"Name":<10} {"Math":>5} {"Science":>8} {"English":>8} {"Avg":>6}')
    print('-' * 45)
    for row in reader:             # iterate over each student row
        name    = row['Name']      # student name
        math    = int(row['Math'])      # convert string to int
        science = int(row['Science'])
        english = int(row['English'])
        avg = (math + science + english) / 3   # arithmetic mean
        print(f'{name:<10} {math:>5} {science:>8} {english:>8} {avg:>6.1f}')
```

```
import json   # Python's built-in JSON module

# ── Saving and loading data with JSON ────────────────────────────────
math_constants = {
    'pi'    : 3.14159265,
    'e'     : 2.71828182,
    'phi'   : 1.61803398,
    'sqrt2' : 1.41421356
}

# Write to JSON file
with open('constants.json', 'w') as f:
    json.dump(math_constants, f, indent=4)   # indent=4 makes it human-readable

# Read back from JSON file
with open('constants.json', 'r') as f:
    loaded = json.load(f)   # parse JSON back to Python dictionary

print('Loaded from JSON:')
for key, val in loaded.items():
    print(f'  {key:8s} = {val}')
```

### 💡 Key Takeaway
File handling lets you persist and share mathematical data. `.csv` is the universal format for tabular/numerical data. `.json` is ideal for structured configuration and results. Always use `with open()` to ensure files are properly closed.
