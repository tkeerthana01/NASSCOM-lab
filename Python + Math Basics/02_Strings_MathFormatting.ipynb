# 02 – Strings & Mathematical Formatting

## 📖 Theory

A **string** is a sequence of characters enclosed in single `' '` or double `" "` quotes.

In mathematics we often need to **display results clearly**. Python provides powerful tools:
- **f-strings** (`f'...'`) – embed expressions directly inside a string
- **format()** – control decimal places, width, and alignment
- **String methods** – `upper()`, `lower()`, `replace()`, `split()`, etc.

Good formatting makes mathematical outputs readable and professional.

```
# ── Creating strings ─────────────────────────────────────────────────
subject  = 'Mathematics'   # string using single quotes
greeting = "Hello World"   # string using double quotes

print(subject)    # Mathematics
print(greeting)   # Hello World
```

```
# ── String operations ────────────────────────────────────────────────
word = 'python'

print(word.upper())          # PYTHON  – converts to uppercase
print(word.capitalize())     # Python  – capitalises first letter
print(len(word))             # 6       – number of characters
print(word.replace('p','P')) # Python  – replaces character
print(word[0])               # p       – indexing: first character
print(word[-1])              # n       – negative index: last character
print(word[1:4])             # yth     – slicing: characters 1 to 3
```

```
# ── f-strings for mathematical results ──────────────────────────────
radius = 7.0                      # radius of a circle
pi     = 3.14159265358979         # value of π

area      = pi * radius ** 2      # area = π r²
circumference = 2 * pi * radius   # circumference = 2πr

# f-string: variables in curly braces are evaluated automatically
print(f'Radius        = {radius}')
print(f'Area          = {area:.4f}')          # 4 decimal places
print(f'Circumference = {circumference:.4f}') # 4 decimal places
```

```
# ── format() for table-style output ─────────────────────────────────
print('\n--- Multiplication Table for 5 ---')
for i in range(1, 11):                          # loop 1 to 10
    result = 5 * i                              # compute product
    # {:2d} = integer with width 2, {:3d} = integer with width 3
    print('5 × {:2d} = {:3d}'.format(i, result))
```

### 💡 Key Takeaway
f-strings with `:.2f` notation control decimal precision. String formatting is essential when presenting mathematical results in reports or tables.
