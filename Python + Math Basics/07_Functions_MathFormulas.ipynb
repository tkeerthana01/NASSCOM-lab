# 07 – Functions & Mathematical Formulas

## 📖 Theory

A **function** is a reusable block of code that takes **inputs (parameters)**, performs a computation, and returns an **output**.

This maps directly to mathematical functions:
$$f(x) = 2x + 3 \quad \Rightarrow \quad \text{input: } x, \text{ output: } 2x+3$$

### Why use functions?
- **Reusability** – write once, use many times
- **Clarity** – named formulas are easier to understand
- **DRY principle** – Don't Repeat Yourself

```
def function_name(parameter1, parameter2):
    # body
    return result
```

````
# ── Defining and calling a simple function ───────────────────────────
def square(x):          # define function 'square' with parameter x
    result = x ** 2     # compute x squared
    return result       # return the computed value

# Calling the function
print(square(4))        # → 16   (4² = 16)
print(square(9))        # → 81   (9² = 81)
print(square(-3))       # → 9    ((-3)² = 9)
````

````
# ── Functions for geometric formulas ────────────────────────────────
import math   # import Python's math module for π and √

def circle_area(r):
    """Returns area of a circle given radius r: A = πr²"""
    return math.pi * r ** 2       # math.pi gives precise value of π

def circle_circumference(r):
    """Returns circumference: C = 2πr"""
    return 2 * math.pi * r

def triangle_area(base, height):
    """Returns area of a triangle: A = ½ × base × height"""
    return 0.5 * base * height

# Test the functions
r = 5
print(f'Circle  Area (r={r})        : {circle_area(r):.4f}')
print(f'Circle  Circumference (r={r}): {circle_circumference(r):.4f}')
print(f'Triangle Area (b=6, h=4)    : {triangle_area(6, 4):.4f}')
````

````
# ── Default parameter values ─────────────────────────────────────────
def power(base, exponent=2):    # exponent defaults to 2 if not provided
    """Raises base to the given exponent. Default: square."""
    return base ** exponent

print(power(3))       # → 9  (3² — uses default exponent=2)
print(power(3, 3))    # → 27 (3³ — overrides default)
print(power(2, 10))   # → 1024 (2¹⁰)
````

````
# ── Function returning multiple values ───────────────────────────────
def quadratic_roots(a, b, c):
    """Returns both roots of ax² + bx + c = 0 using the quadratic formula."""
    discriminant = b**2 - 4*a*c    # Δ = b² - 4ac
    if discriminant < 0:
        return None, None          # no real roots when Δ < 0
    root1 = (-b + discriminant**0.5) / (2*a)   # (-b + √Δ) / 2a
    root2 = (-b - discriminant**0.5) / (2*a)   # (-b - √Δ) / 2a
    return root1, root2            # return both roots as a tuple

# Solve: x² - 5x + 6 = 0 → roots are 3 and 2
r1, r2 = quadratic_roots(1, -5, 6)
print(f'Roots of x² - 5x + 6 = 0 : x₁ = {r1}, x₂ = {r2}')

# Solve: x² + 1 = 0 → no real roots
r1, r2 = quadratic_roots(1, 0, 1)
print(f'Roots of x² + 1 = 0       : {r1}, {r2} (no real solution)')
````

````
# ── Lambda (anonymous) functions – one-line math ─────────────────────
double  = lambda x: x * 2          # f(x) = 2x
cube    = lambda x: x ** 3         # f(x) = x³
celsius_to_fahrenheit = lambda c: (c * 9/5) + 32  # temperature conversion

print(double(7))                    # 14
print(cube(4))                      # 64
print(celsius_to_fahrenheit(100))   # 212.0 — boiling point
````

### 💡 Key Takeaway
Functions are Python's way to encode mathematical formulas as reusable tools. Default parameters, multiple return values, and lambda functions give immense flexibility.
