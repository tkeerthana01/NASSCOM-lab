# 01 – Variables, Data Types & Basic Arithmetic

## 📖 Theory

A **variable** is a named container that stores a value in memory. Think of it like a labelled box.

Python supports several built-in **data types**:
| Type | Example | Description |
|------|---------|-------------|
| `int` | `5` | Whole numbers |
| `float` | `3.14` | Decimal numbers |
| `str` | `'hello'` | Text |
| `bool` | `True` | True / False |

**Arithmetic operators** in Python mirror standard mathematics:
- `+` Addition, `-` Subtraction, `*` Multiplication, `/` Division
- `**` Exponentiation (power), `//` Floor division, `%` Modulo (remainder)


```
# ── Assigning variables ──────────────────────────────────────────────
x = 10          # integer variable: whole number
y = 3.5         # float variable: decimal number
name = 'Alice'  # string variable: text
is_ready = True # boolean variable: True or False

# ── Checking data types ──────────────────────────────────────────────
print(type(x))         # <class 'int'>
print(type(y))         # <class 'float'>
print(type(name))      # <class 'str'>
print(type(is_ready))  # <class 'bool'>
```

```
# ── Basic arithmetic operations ──────────────────────────────────────
a = 15   # first number
b = 4    # second number

print('Addition       :', a + b)   # 15 + 4 = 19
print('Subtraction    :', a - b)   # 15 - 4 = 11
print('Multiplication :', a * b)   # 15 × 4 = 60
print('Division       :', a / b)   # 15 ÷ 4 = 3.75  (always float)
print('Floor Division :', a // b)  # 15 ÷ 4 = 3     (integer quotient)
print('Modulo         :', a % b)   # remainder of 15 ÷ 4 = 3
print('Exponentiation :', a ** b)  # 15 raised to power 4 = 50625
```

```
# ── Mathematical expressions combining variables ─────────────────────
length = 7.0    # length of a rectangle
breadth = 4.5   # breadth of a rectangle

area      = length * breadth          # area = l × b
perimeter = 2 * (length + breadth)    # perimeter = 2(l + b)

print('Area      :', area)            # 31.5
print('Perimeter :', perimeter)       # 23.0
```

### 💡 Key Takeaway
Variables store values; arithmetic operators let us compute mathematical expressions. Python's `/` always returns a float, while `//` returns the integer quotient.
