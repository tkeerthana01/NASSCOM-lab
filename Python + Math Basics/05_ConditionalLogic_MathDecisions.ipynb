# 05 – Conditional Logic & Mathematical Decisions

## 📖 Theory

**Conditional statements** allow a program to make decisions based on conditions.
In mathematics, we often need piecewise functions like:

$$f(x) = \begin{cases} x^2 & \text{if } x \geq 0 \\ -x & \text{if } x < 0 \end{cases}$$

Python uses `if`, `elif` (else if), and `else` for this.

**Comparison operators:** `==` equal, `!=` not equal, `>`, `<`, `>=`, `<=`

**Logical operators:** `and`, `or`, `not`

```
# ── Simple if-else ────────────────────────────────────────────────────
number = -7   # the number to test

if number > 0:              # condition: is number positive?
    print('Positive number')     # runs if condition is True
elif number == 0:           # else-if: is number zero?
    print('Number is zero')      # runs if this condition is True
else:                       # none of the above matched
    print('Negative number')     # runs if all conditions are False
```

```
# ── Piecewise mathematical function ──────────────────────────────────
# f(x) = x²  if x >= 0
#        -x  if x < 0
def f(x):
    if x >= 0:         # non-negative: use x squared
        return x ** 2
    else:              # negative: use absolute value
        return -x

# Test the function for different values
for val in [-3, -1, 0, 2, 5]:   # loop over test values
    print(f'f({val:3d}) = {f(val)}')
```

```
# ── Grade classification using if-elif-else ──────────────────────────
score = 73   # student's exam score

if score >= 90:               # 90 and above
    grade = 'A (Excellent)'
elif score >= 80:             # 80 to 89
    grade = 'B (Good)'
elif score >= 70:             # 70 to 79
    grade = 'C (Average)'
elif score >= 60:             # 60 to 69
    grade = 'D (Below Average)'
else:                         # below 60
    grade = 'F (Fail)'

print(f'Score: {score} → Grade: {grade}')
```

```
# ── Logical operators: and / or / not ────────────────────────────────
x = 15   # test number

# 'and' – both conditions must be True
if x > 10 and x < 20:
    print(f'{x} is between 10 and 20')   # True and True → prints

# 'or' – at least one condition must be True
if x < 5 or x > 12:
    print(f'{x} is outside the range 5–12')  # False or True → prints

# 'not' – inverts the condition
if not x == 10:
    print(f'{x} is not equal to 10')    # not False → True → prints
```

```
# ── Checking divisibility (modulo + conditionals) ────────────────────
n = 36   # number to test

if n % 2 == 0:          # remainder when divided by 2 is 0?
    print(f'{n} is Even')
else:
    print(f'{n} is Odd')

if n % 3 == 0 and n % 5 == 0:   # divisible by both 3 and 5?
    print(f'{n} is divisible by both 3 and 5')
elif n % 3 == 0:
    print(f'{n} is divisible by 3 only')
elif n % 5 == 0:
    print(f'{n} is divisible by 5 only')
```

### 💡 Key Takeaway
Conditionals are Python's way of expressing piecewise functions and decision rules. Combined with `and`/`or`/`not`, they handle any logical condition in mathematics.
