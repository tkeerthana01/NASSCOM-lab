# 12 – Object-Oriented Programming & Math Classes

## 📖 Theory

**Object-Oriented Programming (OOP)** organises code into **classes** and **objects**.

- A **class** is a blueprint (like the definition of 'fraction' in mathematics)
- An **object (instance)** is a specific example (like the fraction 3/4)

### Key OOP concepts:
| Concept | Meaning |
|---------|--------|
| **Class** | Template / blueprint |
| **Object** | Specific instance of a class |
| **Attribute** | Data stored in the object |
| **Method** | Function that belongs to the object |
| **`__init__`** | Constructor – runs when object is created |

OOP is perfect for mathematical entities like fractions, vectors, shapes, and polynomials.

```
# ── Defining a simple class: Shape ───────────────────────────────────
class Rectangle:
    """Represents a rectangle with length and breadth."""

    def __init__(self, length, breadth):   # constructor method
        self.length  = length              # attribute: store length
        self.breadth = breadth             # attribute: store breadth

    def area(self):                        # method: compute area
        return self.length * self.breadth  # A = l × b

    def perimeter(self):                   # method: compute perimeter
        return 2 * (self.length + self.breadth)  # P = 2(l+b)

    def is_square(self):                   # method: check if square
        return self.length == self.breadth # True if sides are equal

# ── Creating objects (instances) ─────────────────────────────────────
r1 = Rectangle(8, 5)    # object r1: length=8, breadth=5
r2 = Rectangle(6, 6)    # object r2: a square

print(f'r1 Area      : {r1.area()}')
print(f'r1 Perimeter : {r1.perimeter()}')
print(f'r1 Is Square : {r1.is_square()}')
print(f'r2 Is Square : {r2.is_square()}')
```

```
# ── A Fraction class with mathematical operations ─────────────────────
import math

class Fraction:
    """Represents a mathematical fraction p/q."""

    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError('Denominator cannot be zero!')  # guard clause
        gcd = math.gcd(numerator, denominator)    # simplify using GCD
        self.num = numerator   // gcd              # simplified numerator
        self.den = denominator // gcd              # simplified denominator

    def __str__(self):             # string representation
        return f'{self.num}/{self.den}'

    def add(self, other):          # addition: a/b + c/d = (ad+bc)/(bd)
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)   # return new Fraction (auto-simplifies)

    def multiply(self, other):     # multiplication: (a/b)×(c/d) = ac/bd
        return Fraction(self.num * other.num, self.den * other.den)

# Test the Fraction class
f1 = Fraction(1, 2)    # represents ½
f2 = Fraction(2, 3)    # represents ⅔

print(f'{f1} + {f2} = {f1.add(f2)}')       # ½ + ⅔ = 7/6
print(f'{f1} × {f2} = {f1.multiply(f2)}')  # ½ × ⅔ = 1/3
```

```
# ── A Vector2D class ─────────────────────────────────────────────────
import math

class Vector2D:
    """Represents a 2D vector (x, y)."""

    def __init__(self, x, y):
        self.x = x   # x-component
        self.y = y   # y-component

    def __str__(self):
        return f'({self.x}, {self.y})'

    def magnitude(self):              # |v| = √(x² + y²)
        return math.sqrt(self.x**2 + self.y**2)

    def add(self, other):             # vector addition: (a+c, b+d)
        return Vector2D(self.x + other.x, self.y + other.y)

    def dot(self, other):             # dot product: ax·bx + ay·by
        return self.x * other.x + self.y * other.y

    def scale(self, scalar):          # scalar multiplication
        return Vector2D(self.x * scalar, self.y * scalar)

v1 = Vector2D(3, 4)
v2 = Vector2D(1, 2)

print(f'v1             = {v1}')
print(f'|v1|           = {v1.magnitude()}')           # 5.0
print(f'v1 + v2        = {v1.add(v2)}')               # (4, 6)
print(f'v1 · v2        = {v1.dot(v2)}')               # dot product = 11
print(f'v1 × 2         = {v1.scale(2)}')              # (6, 8)
```

### 💡 Key Takeaway
Classes let you model mathematical objects (fractions, vectors, shapes) with their own operations. The `__init__` constructor initialises attributes; methods encode mathematical operations. OOP makes complex mathematical code organised and reusable.
