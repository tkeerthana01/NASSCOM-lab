# 04 – Dictionaries & Sets

## 📖 Theory

### Dictionary
A **dictionary** stores data as **key–value pairs** (like a real dictionary: word → definition).
Enclosed in `{ }`, each entry looks like `'key': value`.

In mathematics, dictionaries are useful to map variable names to values, store formula constants, or label datasets.

### Set
A **set** is an **unordered collection of unique elements**. It directly mirrors the mathematical concept of a **set** from Set Theory.
Operations: Union `|`, Intersection `&`, Difference `-`

```
# ── Creating a dictionary ────────────────────────────────────────────
student = {
    'name'   : 'Priya',     # key 'name' maps to value 'Priya'
    'age'    : 22,           # key 'age'  maps to integer 22
    'grade'  : 'A',          # key 'grade' maps to string 'A'
    'score'  : 94.5          # key 'score' maps to float 94.5
}

print(student)               # prints the full dictionary
print(student['name'])       # access value by key → 'Priya'
print(student['score'])      # access score → 94.5
```

```
# ── Mathematical constants stored in a dictionary ────────────────────
constants = {
    'pi'         : 3.14159265,   # ratio of circumference to diameter
    'e'          : 2.71828182,   # Euler's number (base of natural log)
    'phi'        : 1.61803398,   # golden ratio
    'sqrt2'      : 1.41421356,   # square root of 2
}

print('π  =', constants['pi'])        # 3.14159265
print('e  =', constants['e'])         # 2.71828182
print('φ  =', constants['phi'])       # 1.61803398
print('√2 =', constants['sqrt2'])     # 1.41421356
```

```
# ── Adding, updating, and removing entries ───────────────────────────
marks = {'Math': 88, 'Physics': 75, 'Chemistry': 82}  # initial marks

marks['Biology'] = 91           # add a new subject
marks['Math']    = 95           # update existing value
del marks['Chemistry']          # remove a key

print('Updated marks:', marks)  # {'Math': 95, 'Physics': 75, 'Biology': 91}

# Iterating over key-value pairs
for subject, score in marks.items():       # .items() gives (key, value) pairs
    print(f'  {subject:10s}: {score}')     # formatted output
```

```
# ── Sets – unique elements ────────────────────────────────────────────
A = {1, 2, 3, 4, 5}     # set A
B = {3, 4, 5, 6, 7}     # set B

print('A           :', A)
print('B           :', B)
print('Union A|B   :', A | B)   # all elements in A or B (no duplicates)
print('Intersection:', A & B)   # elements common to both A and B → {3,4,5}
print('Difference  :', A - B)   # elements in A but NOT in B → {1,2}
print('Symmetric Δ :', A ^ B)   # elements in A or B but NOT both → {1,2,6,7}
```

```
# ── Sets remove duplicates automatically ─────────────────────────────
numbers_with_duplicates = [1, 2, 2, 3, 3, 3, 4, 5, 5]  # raw list
unique_numbers = set(numbers_with_duplicates)            # convert to set

print('Original :', numbers_with_duplicates)   # has duplicates
print('Unique   :', unique_numbers)            # only unique values
```

### 💡 Key Takeaway
Dictionaries map names to values — ideal for formulas and labelled data. Sets embody mathematical Set Theory with Union, Intersection, and Difference operations.
