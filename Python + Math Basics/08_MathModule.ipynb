# 08 – The `math` Module

## 📖 Theory

Python's built-in **`math` module** provides functions and constants for standard mathematical operations.

You import it with `import math` and access items as `math.function_name()`.

### Key categories:
| Category | Functions |
|----------|----------|
| Constants | `math.pi`, `math.e`, `math.inf` |
| Rounding | `math.floor()`, `math.ceil()`, `math.trunc()` |
| Powers & logs | `math.sqrt()`, `math.pow()`, `math.log()`, `math.log10()` |
| Trigonometry | `math.sin()`, `math.cos()`, `math.tan()` |
| Combinatorics | `math.factorial()`, `math.comb()`, `math.perm()` |

```
import math   # import the entire math module

# ── Mathematical constants ────────────────────────────────────────────
print('π  =', math.pi)      # 3.141592653589793
print('e  =', math.e)       # 2.718281828459045 (Euler's number)
print('τ  =', math.tau)     # 6.283… = 2π
print('∞  =', math.inf)     # positive infinity
```

```
# ── Powers and roots ─────────────────────────────────────────────────
print('√16     =', math.sqrt(16))       # square root → 4.0
print('∛27     =', 27 ** (1/3))         # cube root using ** (1/3)
print('2^8     =', math.pow(2, 8))      # 2 to the power 8 → 256.0
print('|−5|    =', math.fabs(-5))       # absolute value → 5.0
```

```
# ── Logarithms ───────────────────────────────────────────────────────
# log₂(8) = 3  because 2³ = 8
# log₁₀(1000) = 3  because 10³ = 1000
# ln(e) = 1  because e¹ = e

print('log₂(8)   =', math.log2(8))        # logarithm base 2
print('log₁₀(100)=', math.log10(100))     # logarithm base 10
print('ln(e)     =', math.log(math.e))    # natural log (base e) → 1.0
print('log₅(125) =', math.log(125, 5))    # log base 5 → 3.0
```

```
# ── Rounding functions ────────────────────────────────────────────────
x = 4.7

print('floor (4.7) =', math.floor(x))  # largest integer ≤ x → 4
print('ceil  (4.7) =', math.ceil(x))   # smallest integer ≥ x → 5
print('trunc (4.7) =', math.trunc(x))  # removes decimal part → 4
print('round (4.7) =', round(x))       # standard rounding → 5
```

```
# ── Trigonometry (angles in RADIANS) ─────────────────────────────────
# Convert degrees to radians: radians = degrees × π / 180
deg = 45
rad = math.radians(deg)   # convert 45° to radians

print(f'Angle: {deg}° = {rad:.6f} radians')
print(f'sin(45°) = {math.sin(rad):.6f}')    # ≈ 0.707107
print(f'cos(45°) = {math.cos(rad):.6f}')    # ≈ 0.707107
print(f'tan(45°) = {math.tan(rad):.6f}')    # ≈ 1.000000

# Pythagorean identity: sin²θ + cos²θ = 1
identity = math.sin(rad)**2 + math.cos(rad)**2
print(f'sin²(45°) + cos²(45°) = {identity:.6f}')  # should be 1.0
```

```
# ── Combinatorics ────────────────────────────────────────────────────
# Factorial: n! = 1×2×3×…×n
print('5! =', math.factorial(5))     # 120

# Combination: C(n,r) = n! / (r! × (n-r)!)  — choosing r from n
print('C(6,2) =', math.comb(6, 2))   # 15 ways to choose 2 from 6

# Permutation: P(n,r) = n! / (n-r)!  — arranging r from n
print('P(6,2) =', math.perm(6, 2))   # 30 arrangements of 2 from 6
```

### 💡 Key Takeaway
The `math` module covers virtually every standard mathematical operation. Always convert degrees to radians for trigonometric functions using `math.radians()`.
