# 10 – Statistics & Probability

## 📖 Theory

**Statistics** is the science of collecting, analysing, and interpreting data.

### Measures of Central Tendency
| Measure | Definition | Formula |
|---------|------------|--------|
| Mean | Average | $\bar{x} = \frac{\sum x_i}{n}$ |
| Median | Middle value | Sort, take centre |
| Mode | Most frequent | Highest frequency |

### Measures of Spread
| Measure | Formula |
|---------|--------|
| Range | max − min |
| Variance | $\sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}$ |
| Std Deviation | $\sigma = \sqrt{\text{Variance}}$ |

**Probability**: P(event) = favourable outcomes / total outcomes

```
import statistics   # Python's built-in statistics module

# ── Dataset ───────────────────────────────────────────────────────────
scores = [72, 85, 90, 68, 74, 85, 92, 78, 85, 88, 63, 91]

# ── Measures of central tendency ─────────────────────────────────────
mean   = statistics.mean(scores)     # arithmetic average
median = statistics.median(scores)   # middle value when sorted
mode   = statistics.mode(scores)     # most frequently occurring value

print(f'Data   : {sorted(scores)}')
print(f'Mean   : {mean:.2f}')        # average of all scores
print(f'Median : {median}')          # middle value
print(f'Mode   : {mode}')            # most common score
```

```
# ── Measures of spread ───────────────────────────────────────────────
import statistics
scores = [72, 85, 90, 68, 74, 85, 92, 78, 85, 88, 63, 91]

data_range  = max(scores) - min(scores)          # spread of data
variance    = statistics.variance(scores)        # avg squared deviation
std_dev     = statistics.stdev(scores)           # square root of variance

print(f'Range      : {data_range}')              # 92 - 63 = 29
print(f'Variance   : {variance:.2f}')            # ~86.42
print(f'Std Dev    : {std_dev:.2f}')             # ~9.30
```

```
# ── Manual variance calculation to understand the formula ────────────
data = [2, 4, 4, 4, 5, 5, 7, 9]   # classic textbook example
n    = len(data)

mean_val = sum(data) / n           # step 1: compute mean
print(f'Mean = {mean_val}')

deviations        = [x - mean_val for x in data]          # step 2: deviation from mean
squared_devs      = [d ** 2 for d in deviations]          # step 3: square each deviation
variance_manual   = sum(squared_devs) / n                  # step 4: average of squares
std_dev_manual    = variance_manual ** 0.5                  # step 5: square root

print(f'Deviations  : {deviations}')
print(f'Sq.Devs     : {squared_devs}')
print(f'Variance    : {variance_manual:.2f}')    # 4.0
print(f'Std Dev     : {std_dev_manual:.2f}')     # 2.0
```

```
# ── Basic Probability ────────────────────────────────────────────────
# Experiment: rolling a fair 6-sided die
total_outcomes = 6    # possible results: 1, 2, 3, 4, 5, 6

# Probability of rolling a 4
p_roll_4 = 1 / total_outcomes
print(f'P(rolling 4)         = {p_roll_4:.4f}')  # 0.1667

# Probability of rolling an even number (2, 4, 6)
even_outcomes = 3     # 2, 4, 6
p_even = even_outcomes / total_outcomes
print(f'P(rolling even)      = {p_even:.4f}')    # 0.5

# Complement rule: P(odd) = 1 - P(even)
p_odd = 1 - p_even
print(f'P(rolling odd)       = {p_odd:.4f}')     # 0.5

# P(rolling < 3) = P(1) + P(2) = 2/6
p_less_3 = 2 / total_outcomes
print(f'P(rolling less than 3) = {p_less_3:.4f}')  # 0.3333
```

```
# ── Simulating probability with random numbers ────────────────────────
import random

random.seed(42)    # seed ensures reproducibility
n_rolls = 10000    # simulate 10,000 dice rolls

count_even = 0     # counter for even results

for _ in range(n_rolls):             # repeat n_rolls times
    roll = random.randint(1, 6)      # random integer from 1 to 6 inclusive
    if roll % 2 == 0:                # check if even
        count_even += 1              # increment counter

experimental_probability = count_even / n_rolls
print(f'Rolls        : {n_rolls}')
print(f'Even count   : {count_even}')
print(f'Experimental P(even) = {experimental_probability:.4f}')  # ≈ 0.5
print(f'Theoretical  P(even) = 0.5000')
```

### 💡 Key Takeaway
The `statistics` module computes mean, median, mode, variance, and std dev directly. Probability = favourable / total. Simulation (Monte Carlo) verifies theoretical probability as n → ∞.
