# 15 – Matplotlib – Visualising Mathematics

## 📖 Theory

**Matplotlib** is Python's fundamental plotting library. Visualisation is essential in mathematics because:
- Graphs reveal patterns and trends invisible in raw numbers
- Plotting functions shows their behaviour across a domain
- Statistical charts communicate distributions and comparisons

### Chart types covered:
| Chart | Use |
|-------|----|
| Line plot | Plotting mathematical functions f(x) |
| Scatter | Relationships between two variables |
| Bar chart | Comparing categories |
| Histogram | Distribution of data |
| Pie chart | Proportions |

```
import matplotlib.pyplot as plt   # standard import alias for matplotlib
import numpy as np                 # for creating x-value arrays

# ── Line plot: f(x) = x² ─────────────────────────────────────────────
x = np.linspace(-4, 4, 200)   # 200 x-values evenly spaced from -4 to 4
y = x ** 2                     # compute y = x² for each x

plt.figure(figsize=(7, 4))     # set figure size (width=7, height=4 inches)
plt.plot(x, y, color='blue', linewidth=2, label='y = x²')  # plot the line
plt.title('Quadratic Function: y = x²')  # chart title
plt.xlabel('x')                           # x-axis label
plt.ylabel('y')                           # y-axis label
plt.axhline(0, color='black', linewidth=0.8)  # horizontal axis line
plt.axvline(0, color='black', linewidth=0.8)  # vertical axis line
plt.grid(True, linestyle='--', alpha=0.6)     # dotted grid
plt.legend()                              # show the legend
plt.tight_layout()                        # prevent label clipping
plt.show()                                # display the plot
```

```
import matplotlib.pyplot as plt
import numpy as np

# ── Multiple functions on one plot ───────────────────────────────────
x = np.linspace(-3, 3, 300)   # x from -3 to 3, 300 points

y1 = x          # f(x) = x     (linear)
y2 = x ** 2     # f(x) = x²    (quadratic)
y3 = x ** 3     # f(x) = x³    (cubic)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='f(x) = x',   color='green',  linewidth=2)
plt.plot(x, y2, label='f(x) = x²',  color='blue',   linewidth=2)
plt.plot(x, y3, label='f(x) = x³',  color='red',    linewidth=2)
plt.title('Linear, Quadratic, and Cubic Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.ylim(-10, 10)   # limit y-axis range for better visibility
plt.grid(True, alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# ── Trigonometric functions ───────────────────────────────────────────
x = np.linspace(0, 2 * np.pi, 300)   # 0 to 2π (one full cycle)

sin_x = np.sin(x)    # sine wave
cos_x = np.cos(x)    # cosine wave

plt.figure(figsize=(9, 4))
plt.plot(x, sin_x, label='sin(x)', color='blue',  linewidth=2)
plt.plot(x, cos_x, label='cos(x)', color='orange', linewidth=2)
plt.title('Trigonometric Functions: sin(x) and cos(x)')
plt.xlabel('x (radians)')
plt.ylabel('Value')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', 'π/2', 'π', '3π/2', '2π'])  # label in terms of π
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(True, alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt

# ── Bar chart: comparing student scores ──────────────────────────────
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']  # categories
scores   = [88,     75,        82,          91,        78]          # values
colors   = ['steelblue', 'tomato', 'gold', 'mediumseagreen', 'orchid']  # bar colors

plt.figure(figsize=(8, 5))
bars = plt.bar(subjects, scores, color=colors, edgecolor='black', width=0.6)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()          # get the bar's y-value
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             str(height), ha='center', va='bottom', fontsize=10)

plt.title('Student Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.ylim(0, 105)     # y-axis from 0 to 105 to fit labels
plt.grid(axis='y', linestyle='--', alpha=0.5)  # horizontal grid lines only
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# ── Histogram: distribution of data ──────────────────────────────────
np.random.seed(42)                          # seed for reproducibility
exam_scores = np.random.normal(70, 12, 200) # 200 scores: mean=70, std=12
exam_scores = np.clip(exam_scores, 0, 100)  # clamp to valid range [0, 100]

plt.figure(figsize=(8, 5))
plt.hist(exam_scores, bins=15, color='steelblue', edgecolor='black', alpha=0.7)
# bins=15: divide data into 15 equal-width bins

# Draw mean and std deviation lines
mean_val = np.mean(exam_scores)
std_val  = np.std(exam_scores)
plt.axvline(mean_val, color='red',    linestyle='--', linewidth=2, label=f'Mean = {mean_val:.1f}')
plt.axvline(mean_val + std_val, color='orange', linestyle=':', linewidth=1.5, label='+1 std')
plt.axvline(mean_val - std_val, color='orange', linestyle=':', linewidth=1.5, label='-1 std')

plt.title('Distribution of Exam Scores (Normal Distribution)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt

# ── Subplots: multiple charts in one figure ───────────────────────────
import numpy as np

x = np.linspace(-3, 3, 200)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))   # 1 row, 3 columns of plots

# Plot 1: y = x²
axes[0].plot(x, x**2, color='blue')
axes[0].set_title('y = x²')
axes[0].set_xlabel('x'); axes[0].set_ylabel('y')
axes[0].grid(True, alpha=0.4)

# Plot 2: y = sin(x)
axes[1].plot(x, np.sin(x), color='orange')
axes[1].set_title('y = sin(x)')
axes[1].set_xlabel('x'); axes[1].set_ylabel('y')
axes[1].grid(True, alpha=0.4)

# Plot 3: y = e^x
axes[2].plot(x, np.exp(x), color='green')
axes[2].set_title('y = eˣ')
axes[2].set_xlabel('x'); axes[2].set_ylabel('y')
axes[2].grid(True, alpha=0.4)

plt.suptitle('Three Important Mathematical Functions', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

### 💡 Key Takeaway
Matplotlib turns mathematical functions into visual insight. Use `np.linspace` to generate x-values, compute y-values vectorially, then `plt.plot()`. Subplots compare multiple functions side by side. Always label axes and add titles for clarity.
