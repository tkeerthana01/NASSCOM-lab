# 14 – Error Handling & Input Validation in Mathematics

## 📖 Theory

Mathematical operations can fail for valid mathematical reasons:
- Division by zero: $\frac{a}{0}$ is undefined
- Square root of negative: $\sqrt{-1}$ is not real
- Logarithm of zero or negative: $\ln(0) = -\infty$

Python uses **exceptions** to handle these errors gracefully.

```
try:
    # code that might fail
except ExceptionType:
    # what to do if it fails
finally:
    # always runs (cleanup)
```

**Common exceptions:** `ZeroDivisionError`, `ValueError`, `TypeError`, `IndexError`

````
# ── Basic try-except ─────────────────────────────────────────────────
# Without error handling – this would CRASH:
# result = 10 / 0   # ZeroDivisionError!

# With error handling:
try:
    numerator   = 10
    denominator = 0
    result = numerator / denominator   # this raises ZeroDivisionError
    print('Result:', result)            # this line is skipped
except ZeroDivisionError:
    print('Error: Cannot divide by zero!')  # caught and handled

print('Program continues normally after the error handler.')
````

````
import math

# ── Handling multiple exception types ────────────────────────────────
def safe_sqrt(x):
    """Returns square root, handles invalid inputs gracefully."""
    try:
        if not isinstance(x, (int, float)):   # check type first
            raise TypeError('Input must be a number')
        result = math.sqrt(x)                  # may raise ValueError if x < 0
        return result
    except ValueError:
        return f'Error: Cannot take sqrt of negative number ({x})'
    except TypeError as e:
        return f'Error: {e}'

# Test with different inputs
print(safe_sqrt(25))      # 5.0          — valid
print(safe_sqrt(0))       # 0.0          — edge case (valid)
print(safe_sqrt(-9))      # Error message— negative
print(safe_sqrt('hello')) # Error message— wrong type
````

````
# ── Raising custom exceptions ─────────────────────────────────────────
def calculate_bmi(weight_kg, height_m):
    """Calculates Body Mass Index: BMI = weight / height²"""
    if weight_kg <= 0:
        raise ValueError(f'Weight must be positive, got {weight_kg}')   # custom message
    if height_m <= 0 or height_m > 3.0:
        raise ValueError(f'Height must be 0–3 metres, got {height_m}')

    bmi = weight_kg / height_m ** 2   # BMI formula
    return round(bmi, 2)

# Test with valid input
try:
    print('BMI:', calculate_bmi(70, 1.75))   # 22.86
    print('BMI:', calculate_bmi(-5, 1.75))   # triggers ValueError
except ValueError as e:
    print(f'Invalid input: {e}')
````

````
# ── Input validation for a math calculator ───────────────────────────
def safe_divide(a, b):
    """Division with full validation."""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError('Both arguments must be numbers')
    if b == 0:
        raise ZeroDivisionError('Denominator cannot be zero')
    return a / b

# Running several tests
test_cases = [(10, 2), (7, 0), ('a', 3), (15, 4)]

for a, b in test_cases:
    try:
        result = safe_divide(a, b)
        print(f'{a} / {b} = {result}')
    except ZeroDivisionError as e:
        print(f'ZeroDivisionError: {e}')
    except TypeError as e:
        print(f'TypeError: {e}')
````

````
# ── try / except / else / finally ────────────────────────────────────
# 'else'    runs ONLY if no exception was raised
# 'finally' ALWAYS runs (used for cleanup)

def log_base(number, base):
    """Compute logarithm with full error handling."""
    try:
        if number <= 0 or base <= 0 or base == 1:
            raise ValueError('log requires number > 0 and base > 0, base ≠ 1')
        result = math.log(number, base)   # compute log
    except ValueError as e:
        print(f'  Error: {e}')            # handle the error
        result = None
    else:
        print(f'  log_{base}({number}) = {result:.4f}')  # success!
    finally:
        print(f'  [Computed for number={number}, base={base}]')  # always
    return result

import math
print('Test 1:'); log_base(100, 10)   # valid → 2.0
print('Test 2:'); log_base(-5, 10)    # invalid → error
````

### 💡 Key Takeaway
`try/except` makes mathematical programs robust to domain violations (division by zero, negative roots). `raise` lets you enforce mathematical constraints explicitly. `finally` ensures cleanup always happens regardless of success or failure.
